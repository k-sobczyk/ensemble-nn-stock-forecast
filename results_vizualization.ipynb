# Convolutional Neural Network

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_loss_curves(nn_csv_path, mini_csv_path):
    nn_df = pd.read_csv(nn_csv_path)
    mini_df = pd.read_csv(mini_csv_path)

    plt.figure(figsize=(20, 10))
    plt.subplot(1, 1, 1)
    plt.plot(nn_df['Epoch'], nn_df['Train Loss'], label='Train Loss', marker='o')
    plt.plot(nn_df['Epoch'], nn_df['Test Loss'], label='Test Loss', marker='o')
    plt.title('Loss Curve for nn_df', fontsize=20)
    plt.xlabel('Epoch', fontsize=16)
    plt.ylabel('Loss', fontsize=16)
    plt.legend(fontsize=14)
    plt.grid(True)
    plt.gcf().patch.set_facecolor('white')
    plt.show()

    # Plot the loss curves for mini_df
    plt.figure(figsize=(20, 10))
    plt.subplot(1, 1, 1)
    plt.plot(mini_df['Epoch'], mini_df['Train Loss'], label='Train Loss', marker='o')
    plt.plot(mini_df['Epoch'], mini_df['Test Loss'], label='Test Loss', marker='o')
    plt.title('Loss Curve for mini_df', fontsize=20)
    plt.xlabel('Epoch', fontsize=16)
    plt.ylabel('Loss', fontsize=16)
    plt.legend(fontsize=14)
    plt.grid(True)
    plt.gcf().patch.set_facecolor('white')
    plt.show()

plot_loss_curves('results/cnn_training_results.csv', 'results/mini_cnn_training_results.csv')
```
